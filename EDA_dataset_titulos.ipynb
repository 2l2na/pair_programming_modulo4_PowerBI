```python
# Función para contar valores separados por comas en una columna
def contar_paises_disponibles(df, columna_origen, columna_nueva):
    """
    Crea una nueva columna que cuenta el número de valores separados por comas en la columna origen.

    :param df: DataFrame original
    :param columna_origen: Nombre de la columna donde están los países
    :param columna_nueva: Nombre de la nueva columna que será creada
    :return: DataFrame modificado con la nueva columna
    """
    def contar_paises(paises_str):
        if isinstance(paises_str, str):  # Verifica si es una cadena válida
            return len(paises_str.split(","))
        return 0

    # Crear la nueva columna en el DataFrame
    df[columna_nueva] = df[columna_origen].apply(contar_paises)
    return df


```
